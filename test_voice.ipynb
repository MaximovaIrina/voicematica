```python
import torch
import IPython
```

```python
model, _ = torch.hub.load(repo_or_dir="snakers4/silero-models", model="silero_tts", language="ru", speaker="v3_1_ru")
```

```python
text = """
<speak>
<prosody pitch="x-high">
<break time="1s"/> Всем привет! Сегодня мы поговорим о провальных решениях власти в отношении Украины."<break time="500ms"/> Начинаем... <break time="1500ms"/>Методичка-вопросник рассерженного патриота.
</prosody>
</speak>
"""

sub_audio = model.apply_tts(
                ssml_text=text,
                speaker="eugene",
                sample_rate=48000,
                put_accent=True,
                put_yo=True)
IPython.display.Audio(sub_audio, rate=48000)
```
