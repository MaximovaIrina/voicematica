```python
import torch
import IPython
```

```python
model, _ = torch.hub.load(repo_or_dir="snakers4/silero-models", model="silero_tts", language="ru", speaker="v3_1_ru")
```

```python
text = """
<speak>
<prosody pitch="x-high">
<break time="1s"/>  !          ."<break time="500ms"/> ... <break time="1500ms"/>-  .
</prosody>
</speak>
"""

sub_audio = model.apply_tts(
                ssml_text=text,
                speaker="eugene",
                sample_rate=48000,
                put_accent=True,
                put_yo=True)
IPython.display.Audio(sub_audio, rate=48000)
```

```python
text = """
<speak>
<break time="1s"/>  !<break time="3s" strength="x-weak"/>          ."<break time="500ms" /> ... <break time="1500ms"/>-  .
</speak>
"""

sub_audio = model.apply_tts(
                ssml_text=text,
                speaker="eugene",
                sample_rate=48000,
                put_accent=True,
                put_yo=True)
IPython.display.Audio(sub_audio, rate=48000)
```

```python
from librosa.effects import pitch_shift, time_stretch

IPython.display.Audio(pitch_shift(y=sub_audio.numpy(), sr=48000, n_steps=-2), rate=48000)
```
